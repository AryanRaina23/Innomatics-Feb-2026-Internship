```python
# Problem Statement 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
max_score = max(employees.values())

# Find all employees with highest score
top_performers = [name for name, score in employees.items() if score == max_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

```python
# Problem Statement 2: Search Query Keyword Analysis

import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Count frequency of each word
word_freq = {}
for word in query.split():
    word_freq[word] = word_freq.get(word, 0) + 1

# Filter words appearing more than once
result = {word: count for word, count in word_freq.items() if count > 1}

print(result)
```

```python
# Problem Statement 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Store (index, value) for even readings
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem Statement 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Count domain occurrences
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate percentage usage
total_emails = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem Statement 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average_sales = sum(sales) / len(sales)

# Detect spikes > 30% above average
threshold = average_sales * 1.3

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Problem Statement 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count occurrences
id_count = {}
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1

# Display duplicates
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
