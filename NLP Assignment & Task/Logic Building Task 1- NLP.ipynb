```python
import re
from collections import Counter
```

Task 2:Build Advanced Preprocessing Function

```python
def preprocess_text(text):

  #remove numbers
  text = re.sub(r'\d+','',text)

  #remove extra spaces
  text = re.sub(r'\s+',' ',text)

  #handle repeated char
  text = re.sub(r'(.)\1{2,}', r'\1', text)

  #text to lowercase
  text = text.lower()

  #remove urls and email like pattern
  text = re.sub(r'http\S+|www\S+','',text)
  text = re.sub(r'\S+@\S+','',text)

  #tokenize
  tokens = text.split()

  cleaned_tokens = []
  #remove very short tokens
  for word in tokens:
    if len(word) > 2 or word in ["no","not"]:
      cleaned_tokens.append(word)

  cleaned_sentence = " ".join(cleaned_tokens)

  return cleaned_tokens , cleaned_sentence
```

Task 3: Stress Testing

```python
sentences = [
    "😍🔥",
    "Buy now at www.superdeal.com and save 50%!!!",
    "Email me at rohan123@gmail.com ASAP",
    "Brooooo this movie was littt 🔥🔥",
    "I scored 95 marks in exam!!!",
    "Follow me on https://instagram.com/rohan.dev",
    "Nahhhh this service is terribleeee 😡",
    "LOL that meme was funnyyyy 😂😂",
    "Win $1000 today!!! Click http://winmoney.com",
    "I do not like this product at all"
]
```

```python
all_tokens = []

for text in sentences:

    tokens, clean_sentence = preprocess_text(text)

    print("\nOriginal Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)

    all_tokens.extend(tokens)
```

Task 4: Token Analytics

```python
for text in sentences:

    tokens, clean_sentence = preprocess_text(text)

    total_tokens = len(tokens)

    unique_tokens = len(set(tokens))

    avg_length = sum(len(word) for word in tokens) / total_tokens if total_tokens else 0

    print("\nSentence:", clean_sentence)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length,2))
```

Task 5: Frequency Analysis

```python
counter = Counter(all_tokens)

print("Top 10 most freq words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[:-6:-1])
```

Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):

  all_tokens = []
  clean_sentences = []

  for text in text_list:

    tokens, clean_sentence = preprocess_text(text)

    all_tokens.extend(tokens)
    clean_sentences.append(clean_sentence)

  return {
      "tokens" : all_tokens,
      "sentences" : clean_sentences
  }
```

```python
result = full_pipeline(sentences)
print(result)
     
```

## 1. Difference between "Love" and "love" in NLP

- In case-sensitive NLP:
  - "Love" and "love" are treated as different tokens.
- In case normalization (lowercasing):
  - Both become "love" to reduce vocabulary size.

**Conclusion:**  
Case normalization simplifies models but may remove subtle meaning.

## 2. What happens if stopwords are not removed?

Stopwords are common words like *the, is, in, and*.

### Effects:
- Preserves context
- Increases noise
- Reduces model efficiency
- Important words may get diluted

**Example:**  
"The movie is not good"  
Removing stopwords incorrectly may change meaning.

---

## 3. Scenarios where removing stopwords is harmful

### 1. Sentiment Analysis
- "I do not like this product"
- Removing "not" flips the sentiment incorrectly.

### 2. Question Answering / Chatbots
- "What is the capital of India?"
- Removing stopwords → "capital India" loses meaning.

### Additional examples:
- Legal text: "shall not" is critical
- Medical text: "no evidence of disease"

---

## 4. Difference between Stemming and Lemmatization

| Feature | Stemming | Lemmatization |
|--------|---------|--------------|
| Approach | Rule-based | Dictionary + linguistic analysis |
| Output | May not be real word | Valid word |
| Speed | Fast | Slower |
| Accuracy | Lower | Higher |

### Examples:
- Stemming:
  - "running" → "run"
  - "studies" → "studi" (not a real word)

- Lemmatization:
  - "running" → "run"
  - "studies" → "study" (correct word)

## Summary

- "Love" vs "love" depends on case sensitivity
